# DataFrames and Data Visualization

## DataFrame basics

```python
# Setup
import pandas as pd
import numpy as np
```

```python
# Load example dataset: wine
from sklearn.datasets import load_wine
wine = load_wine()
print(f"wine['data'] = \n{wine['data']}")
print(f"wine['target'] = \n{wine['target']}")
print(f"['class'] + wine['feature_names'] = \n{['class'] + wine['feature_names']}")
```

```python
# could load a dataset from a file instead
df_csv = pd.read_csv("filepath_or_buffer")
```

```python
# Create DataFrame from wine dataset
df = pd.DataFrame(
    data = np.c_[wine['target'], wine['data']],
    columns= ['class'] + wine['feature_names']
)

# Peak at data types
df.info()
```

```python
# Convert 'class' datatype to 'category'
df['class'] = df['class'].astype('int32').astype('category')

# look at info again to confirm category change
df.info()
```

```python
# Take a look at numerical aspects of the data
df.describe().T
```

```python
# Take a look at the data itself
df
```

```python
# Group by 'class' and calculate average 'ash' content
df.groupby('class')['ash'].mean()
```

```python
corr = np.round(df.corr(), decimals=3)

def format_vertical_headers(df):
    """Display a dataframe with vertical column headers"""
    styles = [
#         dict(selector="th", props=[('width', '40px')]),
        dict(
            selector="th.col_heading",
            props=[
                ('writing-mode', 'vertical-rl'),
#                 ('transform', 'rotateZ(-180deg)'), 
                ('height', '200px'),
#                 ('vertical-align', 'top'),
            ]
        )
    ]
    return (df.fillna('').style.set_table_styles(styles))

styler = format_vertical_headers(corr)

styler.background_gradient(cmap='coolwarm')
styler.format("{:.2}")
```
